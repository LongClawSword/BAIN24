Q1. Suppose you have a jar with 4 jelly beans: 2 are strawberry-flavored, 1 is blueberry-flavored, and 1 is cinnamon-flavored. You draw one jelly bean at random from the jar.

    a. What is the sample space for this experiment?

    b. We define event A as the jelly bean drawn is strawberry-flavored and event B as The jelly bean drawn is not cinnamon-flavored. What are the probabilities of events A and B?

    c. Are events A and B mutually exclusive? Why or why not?


```python
# a. Sample space
sample_space = ['Strawberry', 'Strawberry', 'Blueberry', 'Cinnamon']

# b. Probabilities of events A and B
event_a = ['Strawberry', 'Strawberry']
event_b = ['Strawberry', 'Strawberry', 'Blueberry']

probability_a = len(event_a) / len(sample_space)
probability_b = len(event_b) / len(sample_space)

print(f"Probability of event A (strawberry-flavored): {probability_a}")
print(f"Probability of event B (not cinnamon-flavored): {probability_b}")

# c. Mutual exclusivity
# Events A and B are not mutually exclusive because it is possible to draw a strawberry-flavored jelly bean that is also not cinnamon-flavored.
```

 Q2. Previously, we defined a **Python function P** to compute the probability of an event using the naive definition of probability. Generalize that function to compute the probability of events when they are
 not all equally likely. Use this new function to compute the probability of events A and B from the previous exercise.

 Hint: you can pass a third argument with the probability of each event.

```python
def P(S, A, probs=None):
    """
    Computes the probability of an event A in a sample space S.

    Args:
        S: The sample space.
        A: The event.
        probs: A list of probabilities for each element in the sample space. If None,
            all elements are assumed to be equally likely.

    Returns:
        The probability of event A.
    """
    if probs is None:
        probs = [1/len(S)] * len(S)
    else:
        assert len(probs) == len(S)

    if set(A).issubset(set(S)):
        return sum([probs[i] for i in range(len(S)) if S[i] in A])
    else:
        return 0

# Compute the probability of events A and B from the previous exercise.
sample_space = ['Strawberry', 'Strawberry', 'Blueberry', 'Cinnamon']
event_a = ['Strawberry', 'Strawberry']
event_b = ['Strawberry', 'Strawberry', 'Blueberry']

probability_a = P(sample_space, event_a)
probability_b = P(sample_space, event_b)

print(f"Probability of event A (strawberry-flavored): {probability_a}")
print(f"Probability of event B (not cinnamon-flavored): {probability_b}")
```

Q3.  Use PreliZ to explore different parameters for the Gaussian distributions. Use the methods `plot_pdf`, `plot_cdf`, and `plot_interactive`. We discussed the probability mass/density functions and the cumulative density function. But there are other ways to represent functions like the percentile point function ppf. Using the `plot_ppf` method of PreliZ, plot the percentile point function for the BetaBinomial and Gaussian distributions. Can you explain how the ppf is related to the cdf and pmf/pdf?

```python
# Import PreliZ library
import preliz as p

# Define a Gaussian distribution with mean 0 and standard deviation 1
gaussian = p.Normal(mu=0, sigma=1)

# Plot the probability density function (pdf)
gaussian.plot_pdf()

# Plot the cumulative distribution function (cdf)
gaussian.plot_cdf()

# Plot the interactive plot
gaussian.plot_interactive()

# Plot the percentile point function (ppf)
gaussian.plot_ppf()

# Explain the relationship between the ppf, cdf, and pdf
print("The percentile point function (ppf) is the inverse of the cumulative distribution function (cdf).")
print("It gives the value of the random variable that corresponds to a given probability.")
print("The ppf can be used to find the value of the random variable that has a certain probability of being less than or equal to it.")

# For example, if we want to find the value of the random variable that has a 50% probability of being less than or equal to it, we can use the ppf as follows:
print(gaussian.ppf(0.5))

# This will return the value 0, which is the mean of the Gaussian distribution.
betabinomial = p.BetaBinomial(alpha=10, beta=5, n=20)
betabinomial.plot_ppf()
print(betabinomial.ppf(0.5))
print(gaussian.ppf(0.5))
```

Q4. Use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes,
like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. Report your obervations.

```python
# Import necessary libraries
import preliz as pz
import numpy as np

# Define the SkewNormal distribution with different parameters
skew_normal_dist = pz.distributions.SkewNormal(alpha=5, sigma=10, mu=15)

# Compute the moments of the distribution
moments = skew_normal_dist.moments("mv")
mean = moments[0]
variance = moments[1]

# Generate random samples of different sizes
sample_sizes = [10, 100, 1000]
for sample_size in sample_sizes:
  samples = skew_normal_dist.rvs(sample_size)

  # Calculate the sample mean and variance
  sample_mean = np.mean(samples)
  sample_variance = np.var(samples)

  # Print the results
  print(f"Sample size: {sample_size}")
  print(f"True mean: {mean}, Sample mean: {sample_mean}")
  print(f"True variance: {variance}, Sample variance: {sample_variance}")

# Observations:
# As the sample size increases, the sample mean and variance get closer to the true values.
# This is because the larger the sample size, the more likely it is to be representative of the population.
```
